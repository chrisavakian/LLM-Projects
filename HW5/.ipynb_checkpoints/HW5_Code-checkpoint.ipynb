```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Now, let's read and display the first few lines of the Titanic dataset to understand its structure.
titanic_data_path = 'datasets/titanic/titanic_training.csv'

titanic_df = pd.read_csv(titanic_data_path)
titanic_df.head()
```

```python
# Impute missing values for 'age' with median
age_imputer = SimpleImputer(strategy='median')
titanic_df['age'] = age_imputer.fit_transform(titanic_df[['age']])

# For simplicity, drop rows with any remaining missing values (e.g., 'embarked', 'sex')
titanic_df.dropna(subset=['embarked', 'sex'], inplace=True)

# Encode categorical variables ('sex', 'embarked')
label_encoder = LabelEncoder()
titanic_df['sex'] = label_encoder.fit_transform(titanic_df['sex'])
embarked_encoded = pd.get_dummies(titanic_df['embarked'], prefix='embarked')
titanic_df = pd.concat([titanic_df, embarked_encoded], axis=1).drop(['embarked'], axis=1)

# Dropping columns that won't be used in this simple model for clarity ('ticket', 'cabin', etc.)
titanic_df = titanic_df.drop(['ticket', 'cabin'], axis=1)

# Show the modified dataset
titanic_df.head()

class TreeNode:
    """A node in the decision tree."""
    def __init__(self, gini, num_samples, num_samples_per_class, predicted_class):
        self.gini = gini
        self.num_samples = num_samples
        self.num_samples_per_class = num_samples_per_class
        self.predicted_class = predicted_class
        self.feature_index = 0
        self.threshold = 0
        self.left = None
        self.right = None

def gini_score(groups, classes):
    """Calculate the Gini score for a split."""
    n_instances = float(sum([len(group) for group in groups]))
    gini = 0.0
    for group in groups:
        size = float(len(group))
        if size == 0:
            continue
        score = 0.0
        for class_val in classes:
            p = [row[-1] for row in group].count(class_val) / size
            score += p * p
        gini += (1.0 - score) * (size / n_instances)
    return gini

def split_dataset(index, value, dataset):
    """Split a dataset based on an attribute and an attribute value."""
    left, right = list(), list()
    for row in dataset:
        if row[index] < value:
            left.append(row)
        else:
            right.append(row)
    return left, right

def get_split(dataset):
    """Select the best split point for a dataset."""
    class_values = list(set(row[-1] for row in dataset))
    b_index, b_value, b_score, b_groups = 999, 999, 999, None
    for index in range(len(dataset[0])-1):
        for row in dataset:
            groups = split_dataset(index, row[index], dataset)
            gini = gini_score(groups, class_values)
            if gini < b_score:
                b_index, b_value, b_score, b_groups = index, row[index], gini, groups
    return {'index':b_index, 'value':b_value, 'groups':b_groups}

def to_terminal(group):
    """Create a terminal node value."""
    outcomes = [row[-1] for row in group]
    return max(set(outcomes), key=outcomes.count)

def split(node, max_depth, min_size, depth):
    """Create child splits for a node or make terminal."""
    left, right = node['groups']
    del(node['groups'])
    # check for a no split
    if not left or not right:
        node['left'] = node['right'] = to_terminal(left + right)
        return
    # check for max depth
    if depth >= max_depth:
        node['left'], node['right'] = to_terminal(left), to_terminal(right)
        return
    # process left child
    if len(left) <= min_size:
        node['left'] = to_terminal(left)
    else:
        node['left'] = get_split(left)
        split(node['left'], max_depth, min_size, depth+1)
    # process right child
    if len(right) <= min_size:
        node['right'] = to_terminal(right)
    else:
        node['right'] = get_split(right)
        split(node['right'], max_depth, min_size, depth+1)

def build_tree(train, max_depth, min_size):
    """Build a decision tree."""
    root = get_split(train)
    split(root, max_depth, min_size, 1)
    return root

def print_tree(node, depth=0):
    """Print a decision tree."""
    if isinstance(node, dict):
        print('%s[X%d < %.3f]' % ((depth*' ', (node['index']+1), node['value'])))
        print_tree(node['left'], depth+1)
        print_tree(node['right'], depth+1)
    else:
        print('%s[%s]' % ((depth*' ', node)))
```
